```python
if ( 'static_store' in locals() ) : 
    del static_store;
    
if ( 'user' in locals() ) : 
    del user;

if ( 'static' in locals() ) : 
    del static;
    
if ( 'line_user' in locals() ) : 
    del line_user;
    
if ( 'find_user' in locals() ) : 
    del find_user;
    
if ( 'get_user' in locals() ) : 
    del get_user;

class static_store() : 
    
    def __init__( self ) : 
        self.user   = [];
        pass;
    
    pass;

static = static_store();

class user() : 
    
    def __init__( self , name )  : 
        self.uname            = name;
        self.uscore           = 0;
        self.win              = False;
        self.start            = False;
        pass;
    
    def start_game( self ) : 
        self.start = True;
        pass;
    
    def toString( self ) : 
        return ( '>> ' + self.uname + ' : ' + str( self.uscore ) );
    
    def check_win( self ) : 
        return self.win;
    
    def is_win( self ) : 
        self.win = True;
    
    def check_start( self ) : 
        return self.start;

    def init( self ) : 
        self.win = False;
        pass;
    
    pass;

line_user = 'Freddie';

ans = -1;

print( 'well down ----- ----- ----- ' );
```

```python
def ceeate_ans() : 
    import random;
    global ans;
    ans = random.randint( 1 , 100 );

def find_user( user_name ) : 
    for i in static.user : 
        if( i.uname == user_name ) : 
            return True;
        pass;
    return False;

def get_user( user_name ) : 
    for i in static.user : 
        if( i.uname == user_name ) : 
            return i;
        pass;
    return None;

msg = '';

while( True ) :

    # ---------------------------------------------  --------------------------------------------- 
    if( msg == '' ) : 
        msg = input( '\ncmd : ' );

    # ---------------------------------------------  --------------------------------------------- 
    if( not find_user( line_user ) and msg not in [ 'play' , 'replay' , 'exit' ] ) : 
        print( '0. replay status : ' , 'game not start ! please type play !' );
        msg='';
        continue;
        pass; # end if ------------------------
        
    # ---------------------------------------------  --------------------------------------------- 
    if( msg == 'play' and not find_user( line_user ) ) : 
        c_user = user( line_user );
        static.user.append( c_user );
        print( '1. show this user : ' , c_user.toString() );
        c_user.start_game();
        ceeate_ans();
        del c_user;
        print( '2. replay status : ' , 'game start !' );
        pass; # end if ------------------------

    # ---------------------------------------------  --------------------------------------------- 
    if( msg == 'play' and get_user( line_user ).check_start() ) : 
        print( '2. replay status : ' , 'game is already start !' );
        pass; # end if ------------------------
    
    # ---------------------------------------------
    if( msg == 'replay' ) : 
        if( find_user( line_user ) ) :
            this_user = get_user( line_user );
            this_user.init();
            pass;
        ceeate_ans();
        print( '3. replay status : ' , 'game is restart !' );
        msg = 'continue';
        continue;
        pass; # end if ------------------------

    # ---------------------------------------------  --------------------------------------------- 
    if( msg == 'exit' ) : 
        if( find_user( line_user ) ) :
            static.user.remove( get_user( line_user ) );
            pass
        line_user = '';
        print( '4. replay status : ' , 'game is exit !' );
        break;
        pass; # end if ------------------------
    
    msg = input( '\nplaying : ' );
    
    if( msg not in [ 'play' , 'replay' , 'exit' ] and find_user( line_user ) and get_user( line_user ).check_start() ) : 

        this_user = get_user( line_user );
        
        if( not this_user.check_win() ) : 

            # user turn to guest
            
            if( msg == 'exit' ) : 
                break;
                pass;
                
            if( msg == 'replay' ) : 
                if( find_user( line_user ) ) :
                    this_user = get_user( line_user );
                    this_user.init();
                    pass;
                ceeate_ans();
                print( '5. replay status : ' , 'game is restart !' );
                msg = 'continue';
                continue;
                pass; # end if ------------------------
            
            try : 
                eval( msg );
                pass;
            except : 
                print( '6. replay status : ' , 'input type error please type a number !' );
                msg = 'continue';
                pass;
            
            if( eval( msg ) > ans ) : 
                text = 'too large , please guess lesser !;'
            
            if( eval( msg ) < ans ) : 
                text = 'too less , please guess larger !;'
                
            if( eval( msg ) == ans ) : 
                text = 'Bingo !;'
                this_user.is_win()
                
            if( this_user.check_win() ) : 
                this_user.init();
                this_user.uscore += 10;
                ceeate_ans();
                print( 'W. replay status : ' , 'you win this game ! score : ' + str( this_user.uscore ) );
                print( 'W. replay status : ' , 'this game will auto restart !' );
                continue;
                pass; # end if --------------------------

            else :
                print( 'C. replay status : ' , 'your this turn status ! ' + text );
                pass; # end else ------------------------

        pass; # end if --------------------------
    
    pass; # end if --------------------------
```
